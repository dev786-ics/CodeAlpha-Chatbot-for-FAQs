```python
import json

with open(r"C:\Users\anjan\Desktop\work_space\rag_work\CodeAlpha-Chatbot-for-FAQs\ai_ml_faq.json") as f:
    data = json.load(f) 
    
```

```python
# text cleaning using stopword
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = re.sub(r'http\S+|www\S+', '', text.lower())
    text = re.sub(r'[^a-z\s]', '', text)
    return " ".join(w for w in text.split() if w not in stop_words)
    
```

```python
print(type(data))
```

```python
processed_data = []

for item in data:
    processed_data.append({
        "question": preprocess(item["question"]),
        "answer": preprocess(item["answer"])
    })
    
```

```python
# convert text to vector
questions = [item["question"] for item in data]
answers = [item["answer"] for item in data]

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(questions)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def chatbot(user_query):
    user_vec = vectorizer.transform([user_query])
    similarity = cosine_similarity(user_vec, X)
    index = similarity.argmax()
    score = similarity[0][index]

    if score < 0.2:
        return "Sorry, I don't understand your question."
    
    return answers[index]   
```

```python
# test using while llop
while True:
    user = input("You: ")
    
    if user.lower() == "exit":
        break
    
    print("Bot:", chatbot(user))
```
